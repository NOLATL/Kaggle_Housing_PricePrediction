# Package Imports

```python
import pandas as pd
from pycaret.regression import *
from google.colab import drive
```

# Reading in Data

```python
drive.mount('/content/drive')

df_train = pd.read_csv('/content/drive/MyDrive/PyRS : Kaggle : Google Gladiators : Housing/train.csv')
df_test = pd.read_csv('/content/drive/MyDrive/PyRS : Kaggle : Google Gladiators : Housing/test.csv')
```

# Linear Regression-Optimized Pycaret Trial

```python
# Setting up linear regression-optimized PyCaret enviro
hs_reg_lopt = setup(df_train, target = 'SalePrice', log_experiment=True, experiment_name="hs_exp_linreg_opt",
               remove_multicollinearity=True, # Removes one feature from each highly-correlated pair
               normalize=True, # standard scaling
               transform_target=True, # Box-cox
               multicollinearity_threshold=0.80, # correlation threshold to define high correlation
               remove_outliers=True,
               outliers_threshold=0.10) # Percentile of top outliers which to remove
```

```python
# Running autoML
best_models_lopt = compare_models(n_select=10, sort='RMSE')
```

```python
# Get the best model
best_model = best_models_lopt[0]
```

```python
# Blending top model
best_10_models = blend_models(best_models_lopt)
```

```python
# Tuning best model
best_model_tuned = tune_model(best_model)
```

```python
best_10_blend_tuned = tune_model(best_10_models)
```

```python
# Finalizing model
final_model = finalize_model(best_model)

# Predicting
predictions = predict_model(final_model, data=df_test)
```

```python
# save_model(best_model_tuned, "bayesian_ridge")
```

```python
# Saving predictions locally
predictions[["Id", "Label"]].rename(columns={"Label":"SalePrice"}).to_csv("predictions_Blend_10_12.01.2021.csv", index=False)
```
